Agrupamento de Dados e Plotagem de Gráficos

```
df2 = df.groupby('Department')['EmployeeCount'].count()
```

```
df2.plot(kind='bar',color='blue')
plt.title('Quantidade de funcionarios por setor')
plt.xlabel('Setor departamento')
plt.ylabel('Quantidade funcionarios')
```

```
df3 = df.groupby('Education')['EmployeeCount'].count()
```

```
df3.plot(kind='line',color='green', marker='o')
plt.title('Quantidade de funcionários níveis de escolaridade')
plt.xlabel('Nivel Escolar')
plt.ylabel('Quantidade funcionarios')
```

```
df4 = df.groupby('Attrition')['EmployeeCount'].count()
```

```
df4.plot(kind='pie',colors=['green','blue'])
plt.title('Funcionarios com atrito ou não')
```
