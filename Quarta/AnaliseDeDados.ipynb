Calculo valor mÃ¡ximo,mÃ­nimo e mÃ©dio das matÃ©rias

```
math = df['math score'].agg(['max','min','mean'])
reading = df['reading score'].agg(['max','min','mean'])
writing = df['writing score'].agg(['max','min','mean'])
print(math) #Matematica nota maxÃ­ma: 100  mÃ­nima: 0 mÃ©dia: 66.089
print(reading) #Leitura nota maxÃ­ma: 100  mÃ­nima: 17 mÃ©dia: 69.169
print(writing) #Escrita nota maxÃ­ma: 100  mÃ­nima: 10 mÃ©dia: 68.054
```

Grafico media de matematica

```
MediaMath = df.groupby('race/ethnicity')['math score'].mean()
print(MediaMath)
```

```
MediaMath.plot(kind='barh')
```

Grafico media de Escrita

```
MediaEscrita = df.groupby('test preparation course')['writing score'].mean()
print(MediaEscrita)
```

```
MediaEscrita.plot(kind='pie', colors=['blue','red'])
plt.title('Media Escrita')
```

Grafico media de leitura

```
MediaLeitura = df.groupby('parental level of education')['reading score'].mean()
print(MediaLeitura)
```

```
MediaLeitura.plot(kind='bar',color='pink')
plt.xticks(rotation=90)
plt.title('Media de Leitura')
```
